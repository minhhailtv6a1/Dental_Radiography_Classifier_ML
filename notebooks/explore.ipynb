# Explore Dental Radiography Classifier

This notebook is intended for exploratory data analysis (EDA) and experimentation with the dental radiography classification model. It will cover the following aspects:

- Loading the dataset
- Visualizing sample images
- Analyzing model predictions
- Evaluating model performance

## Setup

Before running the code, ensure that the necessary libraries are installed and the model is loaded.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import os

# Set visualization style
sns.set(style='whitegrid')

# Load the model and scaler
import joblib
model = joblib.load('../models/lightgbm_model.pkl')
scaler = joblib.load('../models/scaler.pkl')
```

## Load Dataset

Load the dataset containing the images and their corresponding labels for analysis.

```python
# Define the path to the dataset
data_path = '../data/dental_images/'  # Update with the actual path

# Load the dataset
data = pd.read_csv('../data/labels.csv')  # Update with the actual labels file
data.head()
```

## Visualize Sample Images

Display a few sample images from the dataset along with their labels.

```python
# Function to display sample images
def display_sample_images(data, n=5):
    plt.figure(figsize=(15, 5))
    for i in range(n):
        img_path = os.path.join(data_path, data.iloc[i]['filename'])
        img = Image.open(img_path)
        plt.subplot(1, n, i + 1)
        plt.imshow(img)
        plt.title(data.iloc[i]['label'])
        plt.axis('off')
    plt.show()

# Display sample images
display_sample_images(data, n=5)
```

## Analyze Model Predictions

Use the model to make predictions on a subset of the dataset and analyze the results.

```python
# Function to make predictions
def make_predictions(data, model):
    predictions = []
    for i in range(len(data)):
        img_path = os.path.join(data_path, data.iloc[i]['filename'])
        img = Image.open(img_path).resize((224, 224))  # Resize as per model input
        img_array = np.array(img) / 255.0  # Normalize
        img_array = img_array.reshape(1, -1)  # Reshape for model input
        pred = model.predict(img_array)
        predictions.append(pred[0])
    return predictions

# Make predictions on a subset of the data
sample_data = data.sample(10)
sample_data['predictions'] = make_predictions(sample_data, model)
sample_data[['filename', 'label', 'predictions']]
```

## Evaluate Model Performance

Evaluate the model's performance using metrics such as accuracy, confusion matrix, etc.

```python
# Function to evaluate model performance
from sklearn.metrics import accuracy_score, confusion_matrix

def evaluate_model(data):
    y_true = data['label']
    y_pred = data['predictions']
    accuracy = accuracy_score(y_true, y_pred)
    cm = confusion_matrix(y_true, y_pred)
    return accuracy, cm

# Evaluate performance
accuracy, cm = evaluate_model(sample_data)
accuracy, cm
```
